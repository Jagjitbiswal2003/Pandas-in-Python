```python
import pandas as pd
```

```python
data_1 = pd.DataFrame({
        "Days":[1,2,3,4,5,6],
        "Eng":[21,23,45,36,65,32],
        "Math":[32,21,45,65,56,54]
})
print(data_1)
```

```python
# Melt function - vertically print the data
pd.melt(data_1,id_vars=["Days"])
```

```python
# rename variable and value
pd.melt(data_1,id_vars=["Days"],var_name="Data",value_name="data_val")
```

```python
data_1 = pd.DataFrame({
        "Days":[1,2,3,4,5,6],
        "St_name":["a","b","c","a","b","c"],
        "Eng":[21,23,45,36,65,32],
        "Math":[32,21,45,65,56,54]
})
print(data_1)
```

```python
#Pivot() function - it contain  2 args
data_1.pivot(index="Days",columns="St_name")
```

```python
data_1.pivot(index="Days",columns="St_name",values="Eng")
```

```python
data_1.pivot_table(index="St_name",columns="Days",aggfunc="mean")
```
